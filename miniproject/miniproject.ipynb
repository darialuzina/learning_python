```python
df_1 = unique_playstore.copy()
```

```python
df_1['price'] = df_1['price'].str.replace(r'[^\d.]', '', regex=True).astype(float)
```

```python
df_1['price']
```

```python
agg_df = df_1.groupby(
    ['category', 'type']
).agg({
    'price': 'mean',
    'rating':'mean',
    'reviews':'mean'
})
```

```python
rounded_df = agg_df.round({'price': 2, 'rating': 1, 'reviews': 2})
```

```python
rounded_df
```

```python
new_column_names = {
    'price': 'mean_price',
    'rating': 'mean_rating',
    'reviews': 'mean_reviews'
}
```

```python
rounded_df = rounded_df.rename(columns=new_column_names)
```
